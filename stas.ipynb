```python
import pandas as pd 
import numpy as np 
import seaborn as sns
df=pd.read_csv(r'C:\Users\mrdee\OneDrive\Desktop\ds\3-2\week-1\loan_approved.csv')
df
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
#most of the time married people will apply to loan compare to unmarried ones
df['Married'].fillna('Yes',inplace=True)
```

```python
df.isnull().sum()
```

```python
#if one is not married he/she may not have dependents
# If Married == 'No', set Dependents = 0 for missing ones
df.loc[(df['Married'] == 'No') & (df['Dependents'].isna()), 'Dependents'] = '0'

#  For remaining missing Dependents (Married == Yes or Married is NaN)
mode_dep = df['Dependents'].mode()[0]
df['Dependents'].fillna(mode_dep, inplace=True)


```

```python
df.isnull().sum()
```

```python
print(df['Education'].unique())
```

```python
#if one studies well then they sure do job 
#if one doesnt study then there is chance ,they make look after their parents work etc
df.loc[
    (df['Education'] == 'Not Graduate') & (df['Self_Employed'].isna()),
    'Self_Employed'
] = 'Yes'

df.loc[
    (df['Education'] == 'Graduate') & (df['Self_Employed'].isna()),
    'Self_Employed'
] = 'No'
```

```python
df.isnull().sum()
```

```python
print(df['Loan_Status (Approved)'].unique())
```

```python
#high chance is that if loan is approved then one must had previous records of taking loan  
df.loc[(df['Loan_Status (Approved)']=='Y') & (df['Credit_History'].isna()),'Credit_History']='1.0'
df.loc[(df['Loan_Status (Approved)']=='N') & (df['Credit_History'].isna()),'Credit_History']='0.0'
```

```python
df.isnull().sum()
df['Credit_History'].value_counts()
```

```python
df['LoanAmount'].unique()
```

```python
#basic rule
df['LoanAmount'].fillna(0.1*df['ApplicantIncome']+df['CoapplicantIncome'],inplace=True)
```

```python
df.isnull().sum()
```

```python
df['Gender'].unique()
```

```python
df['Gender'].fillna(df['Gender'].mode()[0],inplace=True)
```

```python
def find(x):
    for i in x:
        sai=df[i].skew()
        kai=df[i].kurt()
        print(f"Skewness of {i} : {sai}")
        print(f"kurtosis of {i} : {kai}")
        print('Status : Valid skewness' if sai>-0.5 and sai<0.5 else "Status : Invalid Skewness")
        print('Status : Valid kurtosis' if kai>2 and sai<4 else "Status : Invalid kurtosis")
k=df.columns

find(['ApplicantIncome', 'CoapplicantIncome', 'LoanAmount', 'Loan_Amount_Term'])
```
