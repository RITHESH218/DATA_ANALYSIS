# Seaborn -Data visualization library

### What is seaborn?

Seaborn is a python library that builds on top of matplotlib that makes the statistical data more easy and beutiful

### Why we need it?
1.for creating plots like bio box plot and pair plots.

2.Automatically handles data from pandas dataframes.

### Where we use it?
1.Data analysis and reporting

2.exploring relationships between variables

3.Machine learning for visualizing datasets and results

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df=sns.load_dataset('titanic')
df.head()
```

```python
## 4. Histograms & Countplots
# Histogram
sns.histplot(df['age'].dropna(), bins=10, kde=True)
plt.title("Histogram of Age")
plt.show()
 
# Countplot
sns.countplot(x='class', data=df)
plt.title("Count of Passengers by Class")
plt.show()
 
```

```python
### 5. Scatter Plots
sns.scatterplot(x='age', y='fare', data=df, hue='sex')
plt.title("Age vs Fare")
plt.show()
 
```

```python
### 6. Boxplot & Violin Plot
# Boxplot
sns.boxplot(x='class', y='age', data=df)
plt.title("Boxplot of Age by Class")
plt.show()
 
# Violin plot
sns.violinplot(x='class', y='age', hue='sex', data=df, split=True)
plt.title("Violin Plot of Age by Class and Sex")
plt.show()
 
```

```python
### 7. Pairplot
iris = sns.load_dataset('iris')
sns.pairplot(iris, hue='species')
plt.show()
 
```

```python
### 8. Heatmap
# Correlation heatmap
corr = iris.drop('species', axis=1).corr()
sns.heatmap(corr, annot=True, cmap='gray')
plt.title("Correlation Heatmap of Iris Dataset")
plt.show()
 
```

```python
### 8. Heatmap
# Correlation heatmap (numeric columns only)
corr = iris.select_dtypes(include=['float64', 'int64']).corr()
 
import seaborn as sns
import matplotlib.pyplot as plt
 
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap of Iris Dataset")
plt.show()
 
 
```

```python
### 9. Customization & Themes
sns.set_theme(style="whitegrid")  # options: darkgrid, white, ticks, etc.
sns.histplot(df['age'].dropna(), bins=10, color='orange')
plt.title("Customized Histogram with Seaborn")
plt.show()
 
 
```

```python
sns.barplot(x='class', y='fare', data=df, hue='sex', ci=None)
plt.title("Average Fare by Class and Sex")
plt.show()
 
 
```

```python
"Show trends or changes in mean values across categories."
sns.pointplot(x='class', y='fare', data=df, hue='sex', markers=['o', 'x'])
plt.title("Point Plot of Fare by Class and Sex")
plt.show()
 
 
```

```python
sns.kdeplot(df['fare'].dropna(), fill=True, color='teal')
plt.title("KDE Plot of Fare Distribution")
plt.show()
 
 
```

```python

# 1. Compare Passenger Survival by Class and Gender
sns.barplot(x='class', y='survived', data=df, hue='sex')
plt.title("Passenger Survival by Class and Gender")
plt.show()
```

```python
#2. Distribution of Passenger Ages
sns.histplot(df['age'].dropna(), bins=15, kde=True, color='purple')
plt.title("Distribution of Passenger Ages")
plt.show()
```

```python
#3. Fare Distribution by Class
sns.boxplot(x='class', y='fare', data=df)
plt.title("Fare Distribution by Class")
plt.show()
#4. Relationship Between Age and Fare
sns.scatterplot(x='age', y='fare', data=df, hue='class')
plt.title("Age vs Fare by Class")
plt.show()
```

```python
#5. Correlation Heatmap
corr = df.select_dtypes(include=['float64', 'int64']).corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```
