```python
#most of the time married people will apply to loan compare to unmarried ones
d['Married'].fillna('Married',inplace=True)
```

```python
d.isnull().sum()
```

```python
#if one is not married he/she may not have dependents
# If Married == 'No', set Dependents = 0 for missing ones
d.loc[(d['Married'] == 'No') & (d['Dependents'].isna()), 'Dependents'] = '0'

#  For remaining missing Dependents (Married == Yes or Married is NaN)
mode_dep = d['Dependents'].mode()[0]
d['Dependents'].fillna(mode_dep, inplace=True)

d.isnull().sum()
```

```python
print(d['Education'].unique())
```

```python
#if one studies well then they sure do job 
#if one doesnt study then there is chance ,they make look after their parents work etc
d.loc[
    (d['Education'] == 'Not Graduate') & (d['Self_Employed'].isna()),
    'Self_Employed'
] = 'Yes'

d.loc[
    (d['Education'] == 'Graduate') & (d['Self_Employed'].isna()),
    'Self_Employed'
] = 'No'

d.isnull().sum()
```

```python
print(d['Loan_Status (Approved)'].unique())
```

```python
#high chance is that if loan is approved then one must had previous records of taking loan  
d.loc[(d['Loan_Status (Approved)']=='Y') & (d['Credit_History'].isna()),'Credit_History']='1.0'
d.loc[(d['Loan_Status (Approved)']=='N') & (d['Credit_History'].isna()),'Credit_History']='0.0'
```
